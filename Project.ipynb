```python
# Pollution increase by month distribution
# Count occurrences of each 'Month of Increase'
month_increase_count = df['Month of Increase'].value_counts().reset_index()

# Rename columns for clarity
month_increase_count.columns = ['Month of Increase', 'Occurrences']

# Sort by month order if necessary (ensure the months are ordered correctly)
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
month_increase_count['Month of Increase'] = pd.Categorical(month_increase_count['Month of Increase'], categories=month_order, ordered=True)
month_increase_count = month_increase_count.sort_values('Month of Increase')

# Create the pie chart for pollution increase by month distribution
plt.figure(figsize=(8, 8))

# Create a pie chart for the occurrences of each month
plt.pie(month_increase_count['Occurrences'], labels=month_increase_count['Month of Increase'], autopct='%1.1f%%', startangle=90, colors=sns.color_palette("Set3", len(month_increase_count)))

# Add title
plt.title('Pollution Increase by Month Distribution', fontsize=16)

# Display the plot
plt.show()

```

```python
# Number of cities in maharashtra vs other states
# Group by state and count the number of unique cities in each state
state_city_count = df.groupby('State')['City'].nunique().reset_index()

# Get the count of cities in Maharashtra
maharashtra_city_count = state_city_count[state_city_count['State'] == 'Maharashtra']['City'].values[0]

# Calculate the number of cities in other states
state_city_count['Maharashtra'] = state_city_count['State'].apply(lambda x: state_city_count[state_city_count['State'] == 'Maharashtra']['City'].values[0] if x == 'Maharashtra' else 0)
state_city_count['Other States'] = state_city_count['City'] - state_city_count['Maharashtra']

# Create a stacked bar chart
plt.figure(figsize=(12, 8))

# Plot the stacked bar chart
state_city_count.set_index('State')[['Maharashtra', 'Other States']].plot(kind='bar', stacked=True, color=['#1f77b4', '#ff7f0e'], width=0.8)

# Add title and labels
plt.title('Stacked Bar Chart: Number of Cities in Maharashtra vs Other States', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Number of Cities', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()

```

```python
# Group by 'State' and calculate the average vehicle emissions (%) for each state
state_vehicle_emissions = df.groupby('State')['Vehicle Emissions (%)'].mean().reset_index()

# Find the state with the highest contribution of vehicle emissions
highest_vehicle_emissions_state = state_vehicle_emissions.loc[state_vehicle_emissions['Vehicle Emissions (%)'].idxmax()]

# Print the state with the highest vehicle emissions contribution
print(f"State with the highest contribution of vehicle emissions: {highest_vehicle_emissions_state['State']}")
print(f"Vehicle Emissions Contribution: {highest_vehicle_emissions_state['Vehicle Emissions (%)']}%")

# Visualization: Scatter plot to show vehicle emissions across states
plt.figure(figsize=(10, 6))

# Create a scatter plot
sns.scatterplot(data=state_vehicle_emissions, x='State', y='Vehicle Emissions (%)', color='red', s=100)

# Add title and labels
plt.title('Vehicle Emissions Contribution Across States', fontsize=16)
plt.xlabel('State', fontsize=12)
plt.ylabel('Vehicle Emissions (%)', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Find the correlation between population size and pollutin rate %

# Group by 'State' and calculate the mean of population size and pollution rate (%)
state_pollution_population = df.groupby('State')[['Population', 'Pollution Rate (%)']].mean().reset_index()

# Calculate the correlation between Population and Pollution Rate (%)
correlation_matrix = state_pollution_population[['Population', 'Pollution Rate (%)']].corr()

# Visualization: Heatmap for correlation between population and pollution rate
plt.figure(figsize=(8, 6))

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Add title
plt.title('Correlation Between Population Size and Pollution Rate (%)', fontsize=16)

# Display the plot
plt.tight_layout()
plt.show()

```

```python
# Total deaths due to air pollution in tamil nadu
# 1. Filter the dataset to get data for Tamil Nadu
tamil_nadu_data = df[df['State'] == 'Tamil Nadu']

# 2. Calculate the total number of deaths in Tamil Nadu
total_deaths_tamil_nadu = tamil_nadu_data['Deaths'].sum()

# 3. Create a DataFrame for visualization purposes
death_data = pd.DataFrame({'State': ['Tamil Nadu'], 'Total Deaths': [total_deaths_tamil_nadu]})

# 4. Visualization: Bar chart to show the total deaths in Tamil Nadu
plt.figure(figsize=(8, 6))  # Set figure size for the chart

# Create a bar plot using Seaborn to visualize total deaths in Tamil Nadu
sns.barplot(data=death_data, x='State', y='Total Deaths', hue='State', palette='Reds', legend=False)

# 5. Add title and labels for the chart
plt.title('Total Deaths Due to Air Pollution in Tamil Nadu', fontsize=16)  # Title for the chart
plt.xlabel('State', fontsize=12)  # Label for x-axis
plt.ylabel('Total Deaths', fontsize=12)  # Label for y-axis

# 6. Display the plot
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()  # Show the chart


```

```python
#  find the Agriculture emission vs populattion by state
# 1. Group the dataset by 'State' and calculate the average agricultural emissions for each state
state_agri_emissions = df.groupby('State')['Agricultural Emissions (%)'].mean().reset_index()

# 2. Find the state with the highest agricultural emissions
highest_agri_emissions_state = state_agri_emissions.loc[state_agri_emissions['Agricultural Emissions (%)'].idxmax()]

# Display the state with the highest agricultural emissions
print(f"The state with the highest agricultural emissions is {highest_agri_emissions_state['State']} with {highest_agri_emissions_state['Agricultural Emissions (%)']}% emissions.")

# 3. Create a Bubble Chart: Size of bubbles based on a variable (e.g., Population or Deaths)
# For this example, we'll use 'Population' or 'Deaths' to determine the bubble size
# Let's assume we're using population size for this example
state_population = df.groupby('State')['Population'].sum().reset_index()

# Merging the emissions data with the population data
bubble_data = pd.merge(state_agri_emissions, state_population, on='State')

# 4. Plotting the Bubble Chart
plt.figure(figsize=(10, 6))  # Set figure size for the chart

# Scatter plot to create bubble chart with emissions on x-axis and state on y-axis
plt.scatter(bubble_data['Agricultural Emissions (%)'], bubble_data['State'],
            s=bubble_data['Population'] / 1000000,  # Scale bubble size (divide population for better scaling)
            alpha=0.6, color='orange', edgecolors='w', linewidth=0.5)

# 5. Add titles, labels, and customize the chart
plt.title('Agricultural Emissions vs. Population by State', fontsize=16)
plt.xlabel('Agricultural Emissions (%)', fontsize=12)
plt.ylabel('State', fontsize=12)

# Add a color bar or size legend
plt.scatter([], [], color='orange', alpha=0.6, s=100, label='Bubble size = Population (scaled)')
plt.legend(loc='lower right')

# 6. Display the plot
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()


```
